# TRAVELLING SALESMAN PROBLEM
#### 2 algorithms for initial solution
* **`nearest neighbour`** - greedy algorithm picking nearest neighbour to create path
* **`random`** - random picking nodes to create path

#### 2 aproches to generate candidate path 
* **`swap_many`** - reverses path between two random nodes
* **`swap_one`** - swaps two random nodes in path  

# 1. UNIFORM DISTRIBUTION
### 1) Using random initial solution and swap_many 

```python
N = 100
C = 50
alpha = 0.99998
stop_temp = 0.0000001
stop_iter = 250000
start_temp = 5*C

sa = SimAnn(N, C, start_temp, stop_temp, stop_iter, alpha, generate_random_cities, False)
sa.sim_ann(swap_many)
print("Initial solution")
plot_tour(sa.init_tour, N)
print("Best solution")
plot_tour(sa.best, N)
sa.plot_learning()
sa.plot_temp()
```

### 2) Using nearest neighbour solution and swap_many 

```python
N = 100
C = 50
alpha = 0.99998
stop_temp = 0.0000001
stop_iter = 250000
start_temp = 5*C

sa = SimAnn(N, C, start_temp, stop_temp, stop_iter, alpha, generate_random_cities, True)
sa.sim_ann(swap_many)
print("Initial solution")
plot_tour(sa.init_tour, N)
print("Best solution")
plot_tour(sa.best, N)
sa.plot_learning()
sa.plot_temp()
```
